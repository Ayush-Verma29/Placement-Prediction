```python
#find the q1 and q3 
# if there is point which is less x>q3 -1.5iqr
# and x<q1-1.5iqr
percentile_25=df["degree_p"].quantile(0.25)
percentile_25
```

```python
percentile_75=df["degree_p"].quantile(0.75)
percentile_75
```

```python
iqr=percentile_75-percentile_25
iqr
```

```python
upperlimit=percentile_75+(1.5*iqr)
lowerlimit=percentile_25-(1.5*iqr)
```

```python
upperlimit
```

```python
lowerlimit
```

```python
df=df[df["degree_p"]<=upperlimit]
df=df[df["degree_p"]>=lowerlimit]
```

```python
df
```

```python
sns.boxplot(y="degree_p",data=df)
```

```python
df.columns
```

```python
col=["ssc_p","hsc_p","etest_p","mba_p","salary"]
```

```python
for c in col:
    percentile_25=df[c].quantile(0.25)
    percentile_75=df[c].quantile(0.75)
    iqr=percentile_75-percentile_25
    upperlimit=percentile_75+(1.5*iqr)
    lowerlimit=percentile_25-(1.5*iqr)
    df=df[df[c]<=upperlimit]
    df=df[df[c]>=lowerlimit]
    plt.figure()
    sns.boxplot(y=c,data=df)
```

# K means Clustering 

```python
from sklearn.cluster import KMeans
kmeans=KMeans(6)
```

```python
kmeans.fit(df)
```

```python
kmeans.predict(df)
```

```python
df["clusters"]=kmeans.predict(df)
```

```python
df
```

```python
kmeans.inertia_ # we have nnot used scaling it should be very small 
```

```python
inertia=[]
for i in range(1,11):
    kmeans=KMeans(i)
    kmeans.fit(df.drop(["clusters"],axis=1))
    inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
plt.plot(range(1,11),inertia,marker="o")
plt.xlabel("no of clusters")
plt.ylabel("Inertia")
```
